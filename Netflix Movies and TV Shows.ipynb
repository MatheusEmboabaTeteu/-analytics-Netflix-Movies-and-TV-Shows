```python
plt.figure(figsize=(10,6))
plt.title('Netflix Movies and TV Shows', fontsize=18)
plt.pie(df.type.value_counts(), 
        labels=['Movie', 'Show'],
        textprops={"fontsize":16},
        radius=1.1, startangle=90,
        colors=['#5455be', '#ff8f00'],
        autopct='%.0f%%')
plt.show()
```

```python
sns.catplot(x="type",kind="count",data=df)
```

```python
plt.figure(figsize=(15,6))
plt.title('\nQuantidade de filmes por ano\n', fontsize=18)
sns.countplot(df[df.release_year>1980].release_year, palette='PuRd')
plt.xticks(rotation=90)
plt.xlabel("Anos")
plt.ylabel("Contagem de filmes")
plt.show()
```
